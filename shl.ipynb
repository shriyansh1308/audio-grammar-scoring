```python
import whisper
import language_tool_python
import pandas as pd
import os

# Load models
speech_model = whisper.load_model("base")
tool = language_tool_python.LanguageTool('en-US')

# Paths
train_csv = "/content/drive/MyDrive/shl-audio-scoring-challenge/dataset/csvs/train.csv"
audio_folder = "/content/drive/MyDrive/shl-audio-scoring-challenge/dataset/audios"

train_df = pd.read_csv(train_csv)

results = []

for index, row in train_df.iterrows():
    file_name = row["filename"] + ".wav"
    actual_score = row["label"]

    audio_path = os.path.join(audio_folder, file_name)

    # Speech → Text
    result = speech_model.transcribe(audio_path)
    text = result["text"]

    # Grammar check
    matches = tool.check(text)
    predicted_score = max(10 - len(matches), 0)

    results.append([file_name, actual_score, predicted_score])

df = pd.DataFrame(results, columns=["Audio", "Actual Score", "Predicted Score"])
df.to_csv("train_results.csv", index=False)

print("✅ DONE — Project ready to submit!")
```

```python

!pip install openai-whisper
!pip install language-tool-python

# ================= IMPORTS =================
import whisper
import language_tool_python
import pandas as pd
import os

# ================= LOAD MODELS =================
speech_model = whisper.load_model("base")
tool = language_tool_python.LanguageTool('en-US')

# ================= PATHS =================
train_csv = "/content/drive/MyDrive/shl-audio-scoring-challenge/dataset/csvs/train.csv"
audio_folder = "/content/drive/MyDrive/shl-audio-scoring-challenge/dataset/audios/train"

train_df = pd.read_csv(train_csv)

results = []

# ================= PROCESS TRAIN DATA =================
for index, row in train_df.iterrows():
    file_name = str(row["filename"]).strip() + ".wav"
    actual_score = row["label"]

    audio_path = os.path.join(audio_folder, file_name)

    # Skip if file missing
    if not os.path.exists(audio_path):
        print("Missing:", audio_path)
        continue

    # Speech → Text
    result = speech_model.transcribe(audio_path)
    text = result["text"]

    # Grammar check
    matches = tool.check(text)
    predicted_score = max(10 - len(matches), 0)

    results.append([file_name, actual_score, predicted_score])

# ================= SAVE RESULTS =================
df = pd.DataFrame(results, columns=["Audio", "Actual Score", "Predicted Score"])
df.to_csv("train_results.csv", index=False)

print("🎉 DONE — Train data processed!")

test_csv = "/content/drive/MyDrive/shl-audio-scoring-challenge/dataset/csvs/test.csv"
audio_folder = "/content/drive/MyDrive/shl-audio-scoring-challenge/dataset/audios/test"

test_df = pd.read_csv(test_csv)

test_results = []

for index, row in test_df.iterrows():
    file_name = str(row["filename"]).strip() + ".wav"
    audio_path = os.path.join(audio_folder, file_name)

    if not os.path.exists(audio_path):
        continue

    result = speech_model.transcribe(audio_path)
    text = result["text"]

    matches = tool.check(text)
    predicted_score = max(10 - len(matches), 0)

    test_results.append([file_name, predicted_score])

df_test = pd.DataFrame(test_results, columns=["Audio", "Predicted Score"])
df_test.to_csv("test_predictions.csv", index=False)

print(" Test predictions done!")
```
